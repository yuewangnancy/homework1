# pandas read_html() - Yield Curve Data
Source: US Dept. Of Treasury

#### part A) get all tables from a web page via pandas read_html()

```python
import pandas as pd
```

```python
# use pandas read_html() function to read all tables from a web page
url='https://www.treasury.gov/resource-center/data-chart-center/interest-rates/pages/textview.aspx?data=yield'
dfs = pd.read_html(url)
print(type(dfs))
```

```python
# list of DataFrames are returned from read_html()
for df in dfs:
    print(type(df))
```

#### part B) get one table from a web page via pandas read_html()

```python
# we can target the table of interest using HTML element attributes
# the interest rates <table> has class=t-chart
dfs2 = pd.read_html(url, attrs={"class":"t-chart"})
len(dfs2)
```
